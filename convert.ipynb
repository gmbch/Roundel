```python
from roundel_utils import *
import numpy as np
import nibabel as nib

# Load the .nii file
nii_file = 'data/test/image___pat124_bi.nii.gz'
image = load_nii(nii_file)

mask = load_nii('data/test/masks___pat124_bi.nii.gz')

def remap_labels(frame):
    remapped = np.zeros_like(frame, dtype=np.uint8)
    remapped[frame == 1] = 2
    remapped[frame == 3] = 1
    return remapped

mask = remap_labels(mask)
# Save as NIfTI
nib_mask = nib.Nifti1Image(mask, affine=np.eye(4), dtype='uint8')
nib.save(nib_mask, 'data/test/masks___pat124.nii.gz')
```

```python
from roundel_utils import *
import numpy as np
import nibabel as nib

# Load the .nii file
nii_file = 'data/test/image___patient101_bi.nii.gz'
image = load_nii(nii_file)

mask = np.zeros_like(image, dtype=np.uint8)

# Load specific frames
frame_1 = load_nii('data/test/patient101_frame01_gt.nii.gz')
frame_14 = load_nii('data/test/patient101_frame14_gt.nii.gz')

mask[:, :, :, 0] = frame_1
mask[:, :, :, 13] = frame_14

# Save as NIfTI
nib_mask = nib.Nifti1Image(mask, affine=np.eye(4), dtype='uint8')
nib.save(nib_mask, 'data/test/masks___patient101_bi.nii.gz')
```
